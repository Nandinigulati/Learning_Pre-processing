```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100) #gives 100 values between 0-100
y=np.sin(x)
z=np.cos(x)
```

```python
print(x,y,z)
```

**Plotting the data**

```python
#sine wave
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cos value')
plt.title('cos wave')
plt.show()
```

**BAR PLOT**

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1]) #ax=axis #0,0=coordinates;1,1=height and width
languages=['english','french','spanish','hindi']
people=[100,50,30,150]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('no. of people')
plt.title('language')
plt.show 
```

**PIE CHART**

```python
fig1= plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['english','french','spanish','hindi']
people=[100,50,30,150]
ax.pie(people,labels=languages,autopct='%1.1f%%') #autopct=tells the decimal value
plt.show()
```

**SCATTER PLOT**

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3D SCATTER PLOT
fig3=plt.figure()
ax= plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
